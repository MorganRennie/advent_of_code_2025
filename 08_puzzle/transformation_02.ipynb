```python
import re
import math
input = "input_01.txt"
input_list = []

## separate inputs into a list 
with open(input) as f:
  for line_number, line in enumerate(f):
    line = re.sub(r"\n", " ", line).rstrip()
    line = re.findall(r"\d+", line)
    input_list.append(line)
print(input_list)
```

```python
import math
from itertools import combinations

def dist3d(a, b):
    #print(a[0],b[0])
    set1 = a
    set2 = b
    return math.sqrt((int(set1[0]) - int(set2[0]))**2 + (int(set1[1]) - int(set2[1]))**2 + (int(set1[2]) - int(set2[2]))**2)

def top_10_closest_pairs(points):
    pairs = []

    # combinations() automatically avoids reversed duplicates
    for p1, p2 in combinations(points, 2):
        d = dist3d(p1, p2)
        pairs.append([d, [p1, p2]])

    # sort by distance and take the top 10
    pairs.sort(key=lambda x: x[0])
    return pairs


closest = top_10_closest_pairs(input_list)
pairings = []
for d, a in closest:
    #print(a)
    pairings.append(a)

#print(pairings)
print('complete')
```

```python
#define the empty groups
groups = []

#print(f' starting: groups= {groups}.')

available_links = 10

while available_links > 0:
    for pair in pairings:
        #print(pair)
        print(available_links)
        if available_links <= 0:
            break
        else:
            if groups == []:
                #print(f' there are no groups, starting off we will create the first group. Group list = {group_list}')
                groups = [pair]
                available_links -= 1
            else:
                num_groups = range(len(groups))
                for i in num_groups: 
                    
                    a = pair[0]
                    b = pair[1]
                    group = groups[i]
                   
                    #print(f' looking at pair: {pair} and group: {group})')
                    #print(f'There are {num_groups} total groups. Looking in group {i}: {group}')
                    if a in group and b in group:
                       # print('both values are in the group already, do nothing.')
                        break
                    elif a in group and b not in group:
                        #print(f'{a} is in the group already. add b to the group')
                        groups[i] += [b]
                        available_links -= 1
                        break
                    elif b in group and a not in group:
                        #print(f'{b} is in the group already, add a to the group.')
                        groups[i] += [a]
                        available_links -= 1
                        break
                    elif b not in group and a not in group:
                        #print(f'neither value exists in this group. there are {len(groups)} we are on group {i+1}')
                        if i == len(groups)-1:
                            #print('if this is the last group, create a new group, otherwise continue.')
                            groups.append(pair)
                            available_links -= 1
                            break

print('---------------------')

print(f'There are a total of {len(groups)} groups.')
print(groups)
```

```python
rank1 = 0
rank2 = 0
rank3 = 0

for l in range(len(groups)):
    unique_list = [list(x) for x in set(tuple(x) for x in groups[l])]
    length = len(unique_list) 
    #print(f' for group {l}   here is the unique list: {unique_list}')
    #print(f'    group length is = {len(unique_list)}')

    if length > rank1:
        rank3 = rank2
        rank2 = rank1
        rank1 = length
    elif length > rank2:
        rank3 = rank2
        rank2 = length
    elif length > rank3:
        rank3 = length

print(f' the three top groups are: {rank1}, {rank2}, {rank3}')
final = rank1*rank2*rank3

print(f'final answer is {final}')
```
