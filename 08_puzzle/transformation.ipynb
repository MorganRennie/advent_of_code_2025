```python
import re
import math
input = "input_01.txt"
input_list = []

## separate inputs into a list 
with open(input) as f:
  for line_number, line in enumerate(f):
    line = re.sub(r"\n", " ", line).rstrip()
    line = re.findall(r"\d+", line)
    input_list.append([line])
print(input_list)
```

```python

length = range(len(input_list))

# for each item, find it's closest paring.

closest_item = []
closest_item_distance = 0
master_list = []
pairing_distance = []

for item in length:
    observed_item = input_list[item][0]
    #print(observed_item)
    closest_item = []
    closest_item_distance = 0
    for comp in length: 
        
        observed_comp = input_list[comp][0]
        
        if observed_item != observed_comp:
            set1 = observed_item
            set2 = observed_comp
            distance = math.sqrt((int(set1[0]) - int(set2[0]))**2 + (int(set1[1]) - int(set2[1]))**2 + (int(set1[2]) - int(set2[2]))**2)
            #print(f'for observed location {observed_item}: the distance to the comparisson {observed_comp} is {distance} meters.' )
            
            if closest_item == []:
                #print('there is no current close item.')
                closest_item_distance = distance
                closest_item = [observed_item, observed_comp]
            elif distance < closest_item_distance:
                #print('this distance is less than the current distance')
                closest_item_distance = distance
                closest_item = [observed_item, observed_comp]
                
    pairing_distance += [closest_item,closest_item_distance]
    #print(f'    The closest item for this location {observed_item} is {closest_item}. It is {closest_item_distance} meters away')

    #add all parings into a master list


output = [0] * len(pairing_distance)
for i, (a,b) in enumerate(sorted(range(len(pairing_distance)), key=lambda y: input[y])):
    output[b] = i
```

```python
import math
from itertools import combinations

def dist3d(a, b):
    return math.sqrt(
        (a[0] - b[0])**2 +
        (a[1] - b[1])**2 +
        (a[2] - b[2])**2
    )

def top_10_closest_pairs(points):
    pairs = []

    # combinations() automatically avoids reversed duplicates
    for p1, p2 in combinations(points, 2):
        d = dist3d(p1, p2)
        pairs.append((d, p1, p2))

    # sort by distance and take the top 10
    pairs.sort(key=lambda x: x[0])
    return pairs[:10]


closest = top_10_closest_pairs(points)

for d, a, b in closest:
    print(f"{a} <-> {b}  distance={d:.3f}")
```

```python
# For each pairing, find if they will group.
# check if either of the two items exist in a list, if they do, add it to that grouping. Otherwise, create a new group. 
master_list = stage
group_list = []
left_over_list = master_list
holding = []
#print(left_over_list)

while master_list == left_over_list and master_list != []:
    left_over_list = master_list
    for pairing in master_list:
        #print('--------------')
        holding = []
        #print(f'The masterlist is {len(master_list)} Items long.')
        num_groups = range(len(group_list))
        num_pairs = range(len(pairing))
    
        #print(f' the pairing is {pairing}')
        if group_list == []:
            group_list = [pairing]
            #print(f' there are no groups, starting off we will create the first group. Group list = {group_list}')
            master_list.remove(pairing)
        else:
            for it in num_groups:
                observed_group = group_list[it]
                #print(f' there are {num_groups} groups. Looking at {it}: {observed_group}')
                
                if any(p in observed_group for p in pairing):
                #if set(pairing) & set(observed_group):
                    #print(f' this coord exists in the list {observed_group}')
                    group_list[it] += pairing
                    master_list.remove(pairing)
                    #print(f' the new group list is {group_list}')
                    holding = []
                    break
                else:
                    #print(f' the pairing {pairing} has no values in the group {observed_group}')
                    holding = pairing
            if holding != []:
                #print(f'        Therefore, we add the pairing {holding} as a new group.')
                group_list += [holding]
                master_list.remove(holding)
                holding = []
                #print(f'        The new group list is: {group_list}')
                break
        break

print(f'There are a total of {len(group_list)} groups.')
```
