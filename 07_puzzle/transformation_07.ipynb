```python
## part 1

import re

input = "input_02.txt"

input_list_rows = []
#print(input_list_rows)
beam_split_counter = 0
## separate inputs into a list 
with open(input) as f:
  for line_number, line in enumerate(f):
    line = re.sub(r"\n", " ", line).rstrip()
    input_list_rows.append(line)

#print(input_list_rows)

number_of_rows = range(len(input_list_rows))

#print(number_of_rows)

for y in number_of_rows:
    row = input_list_rows[y]
    previous_row = input_list_rows[y-1]
    splits = 0
    #print(f'row number {y}= {row}')
    number_of_columns = range(len(row))

    for x in number_of_columns:
        row = input_list_rows[y]
        if row[x] == '.':
            if previous_row[x] in ['S','|']:
                input_list_rows[y] = row[:x] + '|' + row[x+1:]
                ##print(row)
            elif (x != 0 and row[x-1]== '^') or (x!= len(row)-1 and row[x+1]== '^'):
                #print(f'Looking at row {row} a period in column {x}. Previous X is "{row[x-1]}". Next X "{row[x+1]}".')
                input_list_rows[y] = row[:x] + '|' + row[x+1:]
                splits += 1
                #print(f'beam is split')
    row = input_list_rows[y]
    number_of_beams_in_row = input_list_rows[y].count('|')
    
    if y != 0:
      numnber_of_beams_in_prev_row = input_list_rows[y-1].count('|')
    else:
      numnber_of_beams_in_prev_row = 0
    if numnber_of_beams_in_prev_row != 0:
      difference = number_of_beams_in_row - numnber_of_beams_in_prev_row
    else: 
       difference = 0
    print(f'row number {y}= {row}. Total Beams: {number_of_beams_in_row}. Split {splits-difference} in this row.')

    beam_split_counter += splits-difference 
print(f'{beam_split_counter} times split')
```
