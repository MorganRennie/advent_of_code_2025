```python
## part 2 -- second attempt.

import re

input = "input_01.txt"

input_list_rows = []
with open(input) as f:
  for line_number, line in enumerate(f):
    line = re.sub(r"\n", " ", line).rstrip()
    input_list_rows.append(line)

#print(input_list_rows)
number_of_rows = len(input_list_rows)
number_of_columns = len(input_list_rows[0])

print(f'width = {number_of_columns}, height = {number_of_rows}')

timeline_grid = [[0] * number_of_columns for _ in range(number_of_rows)]

for y in number_of_rows:
  row = input_list_rows[y]
  print(row)

```

```python
for x, char in enumerate(input_list_rows[0]):
    if char == 'S':
        timeline_grid[0][x] = 1  
        break

for y in range(number_of_rows - 1):  
    for x in range(number_of_columns):
        count = timeline_grid[y][x]
        if count == 0:
            continue
        cell = input_list_rows[y + 1][x]
        if cell == '.':

            timeline_grid[y + 1][x] += count
        elif cell == '^':

            if x > 0:
                timeline_grid[y + 1][x - 1] += count
            if x < number_of_columns - 1:
                timeline_grid[y + 1][x + 1] += count


total_timelines = sum(timeline_grid[-1])
print(f'Total timelines: {total_timelines}')
```
