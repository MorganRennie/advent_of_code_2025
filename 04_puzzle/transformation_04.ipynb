```python
## second stage

import re

input = "input_02.txt"

input_list_rows = []
#print(input_list_rows)

## separate inputs into a list 
with open(input) as f:
  for line_number, line in enumerate(f):
    line = re.sub(r"\n", " ", line).rstrip()
    input_list_rows.append(line)

##print(input_list_rows)
number_of_rows = range(len(input_list_rows))

##print(number_of_rows)

final_available = 0

available = 1
while available != 0:
    available = 0

    number_of_rows = range(len(input_list_rows))

    for y in number_of_rows:
        observed_y = y
        row = input_list_rows[y]
        ##print(f'row number {y}= {row}')
        number_of_columns = range(len(row))
        
        for x in number_of_columns:
            observed_x = x
            rolls = 0
            column = row[x]
            ##print(f'column number {x}= {column}')
    
            a = [observed_x-1, observed_y-1]
            b = [observed_x,   observed_y-1] 
            c = [observed_x+1, observed_y-1]
            
            d = [observed_x-1, observed_y]
            f = [observed_x+1, observed_y] 
            
            g = [observed_x-1, observed_y+1]
            h = [observed_x,   observed_y+1]
            i = [observed_x+1, observed_y+1]
    
            check_coord = [a,b,c,d,f,g,h,i]
    
            for item in check_coord:
                cx, cy = item
                if 0 <= cy < len(input_list_rows) and 0 <= cx < len(input_list_rows[cy]):
                    checked_row = input_list_rows[cy]
                    checked_column = checked_row[cx]
                    ##print(f' for position {item} looking at row: {checked_row} and column value: {checked_column}')
                    if checked_column == '@':
                        rolls += 1
                    ##print(f'total empty spaces: {empty_space}')
                ##else:
                    ##print(f' for position {item} OUT OF BOUNDS')
    
            ##print(f'total rolls: {rolls}')
    
            if rolls < 4 and column == '@': 
                available += 1
                coords = str(observed_x) + ',' + str(observed_y)
                ##print(f'    {coords}')
                ##print(f'    {coords} this roll is available -- total surrpunding rolls: {rolls}')

                input_list_rows[y] = row[:x] + '.' + row[x+1:]
                row = input_list_rows[y]   # refresh local copy
            ##print(f'      Available for moving: {available}')
        ##print(f'           Row Final Available for moving: {available}')
    final_available += available

print(final_available)
```
