```python
## answer for part 1

import re
input = "input_01.txt"
total_combo = 0
with open(input) as f:
  for line_number, line in enumerate(f):
    digit_one = 0
    digit_two = 0
    print('------------------------------')
    print(f'row {line_number}')
    print(f'line {line}')   
    res_str = re.sub(r"\n", "N", line)
    ##print(res_str)
    length = range(len(line)-1)
    for i in length:
        digit = res_str[i]
        try:
            ##print(f'looking at {digit}')
            ##print(f'next digit is {res_str[i+1]}')
            digit = int(digit)
            if digit > digit_one and res_str[i+1]!='N': 
                digit_two = 0
                digit_one = digit
                ##print(f'{digit} is greater than digit one. Updated digit one to {digit_one} and digit two to {digit_two}')
            elif digit > digit_two:
                digit_two = digit
                ##print(f'{digit} is greater than digit two but not one. Updated digit two to {digit_two}')
                highest_combo = str(digit_one) + str(digit_two)
    
        except:
            print(f'{digit} does not work')
    total_combo = total_combo + int(highest_combo)        
    print(f'highest combo here is {digit_one}{digit_two}')
    
    print(f'This brings the total combo to {total_combo}')
print(total_combo)
```
