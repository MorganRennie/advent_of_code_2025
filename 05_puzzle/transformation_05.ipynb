```python
## first stage
import re

input = "input_02.txt"
#print(input)
input_list_rows = []
#print(input_list_rows)

deconstructed = []
collections = []
value = 0
total_value = 0
## separate inputs into a list 
with open(input) as f:
  for line_number, line in enumerate(f):

    search_re = re.compile(r'[^\s]')
    line = re.sub(r"\n", " ", line)
    deconstructed = re.findall(r'[^\s]+', line)
    length = len(deconstructed)
    collections.append(deconstructed)
  
rows = len(collections)-1
print(rows)
for i in range(length):
  operator = collections[rows][i]
  #print(f'for calc number {i}, the operator = {operator}')
  for x in range(len(collections)-1):
    #print(f'   looking at row {x}, the value is {int(collections[x][i])}')
    if x == 0:
      value = int(collections[x][i])
    elif operator == '*':
      value = value * int(collections[x][i])
    elif operator == '+':
      value += int(collections[x][i])
  #print(f'   the final value for calc number {i} is {value}')
  total_value += value

#print('----------------------------------')
print(f'the final value is {total_value}')
```
