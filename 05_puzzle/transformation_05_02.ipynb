```python
## second stage
import re
import math

input = "input_02.txt"
input_list_rows = []
deconstructed = []
collections = []
total_value = 0
list_of_values = []
operator = []
value = 0
length = 0

with open(input) as f:
  for line_number, line in enumerate(f):

    search_re = re.compile(r'[^\s]')
    line = re.sub(r"\n", " ", line)
    deconstructed = re.findall(r'.', line)
    collections.append(deconstructed)

    if len(deconstructed) >= length:
      length = len(deconstructed)
    else: 
      deconstructed += ' '

rows = len(collections)-1



for i in range(length):
  full_digit = ''

  operator = (re.findall(r'[*|+]', collections[rows][i]))

  if operator == ['+'] or operator == ['*']:
    active_operator = operator

  for x in range(rows):
      #print(f'   looking at row {x}, the value is {(collections[x][i])}')
      digit = collections[x][i]
      full_digit = full_digit + digit

  if full_digit == '    ':

    if active_operator == ['*']:
      value = math.prod(list_of_values)
    elif active_operator == ['+']:
      value = sum(list_of_values)
    
    total_value += value
    value = 0
    list_of_values = []
    operator = []
    active_operator = []

  elif  full_digit.strip() !='':
    list_of_values.append(int(full_digit.strip()))

print(total_value)
    
```
